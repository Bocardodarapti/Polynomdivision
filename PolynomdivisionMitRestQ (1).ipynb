Polynomdivision (Division mit Rest) für Polynome über den rationalen Zahlen.

Achtung! Ein Polynom wie x^5+3x^2+4x-5  muss als x **5+3*x**2 + 4*x-5 eingegeben werden.

```python
import sympy as sp

def berechne_polynomdivision(P_in, T_in):
    x = sp.symbols('x')
    
    # Umwandlung der Eingaben in SymPy-Ausdrücke
    P = sp.sympify(P_in)
    T = sp.sympify(T_in)
    
    # Die Division: Q = Quotient, R = Rest
    Q, R = sp.div(P, T)
    
    print(f"Polynom P: {P}")
    print(f"Teiler T:  {T}")
    print("-" * 40)
    print(f"Quotient Q = {Q}")
    print(f"Rest R      = {R}")
    print("-" * 40)
    
    # 1. Die allgemeine Formel
    print("Erinnerung: P = Q * T + R")
    
    # 2. Die Probe mit expliziten Klammern für die Einsetzung
    # Wir nutzen f-Strings, um die Werte in Klammern zu setzen
    probe_einsetzung = f"({Q}) * ({T}) + ({R})"
    ergebnis = sp.simplify(Q * T + R)
    
    print(f"Probe: Q * T + R = {probe_einsetzung} = {ergebnis}")
    
    if ergebnis == P:
        print("Ergebnis: Die Probe ist korrekt!")
```

```python
# --- Eingabebereich ---
P = "x"
T = "1"

berechne_polynomdivision(P, T)
```
